# Label and One Hot Encoding
---

```python
def label_encode(df, cat_cols):
    le = LabelEncoder()
    cat_cols = ["item_type", "item_fat_content", "item_type_2", "outlet_identifier", "outlet_size", "outlet_location_type", "outlet_type"]

    for col in cat_cols:
        df[col] = le.fit_transform(df[col]) 
    
    return df

def onehot_encode(df):
    cat_cols = ["item_type", "item_fat_content", "item_type_2", "outlet_identifier", "outlet_size", "outlet_location_type", "outlet_type"]

    df = label_encode(df, cat_cols)

    df = pd.get_dummies(df, columns=cat_cols[1:])

    return df


```

```python
df = onehot_encode(df)
df.to_csv("data/train_fe.csv")
```
