```python
fig, ax = plt.subplots(1,2, figsize=(25,10))

sns.distplot(df["age"], ax=ax[0]);
ax[0].set_title("Age Distribution")

sns.distplot(df["bmi"], ax=ax[1])
ax[1].set_title("BMI Distribution");

```

```python
df
```

```python
fig, ax = plt.subplots(4,1, figsize=(15,15))

sns.countplot(df["sex"], ax=ax[0]);
ax[0].set_title("Sex Count")

sns.countplot(df["children"], ax=ax[1])
ax[1].set_title("Children Count")

sns.countplot(df["smoker"], ax=ax[2])
ax[2].set_title("Region Count");

sns.countplot(df["region"], ax=ax[3])
ax[3].set_title("Region Count");
```
